```python
%%time

for i in tqdm(range(len(df2)-8)):
    loc = df2['location'][i]
    if (df2['location'][i+8] ==loc):
        df2['RT'][i+8] = (df2['total_cases'][i+7]+df2['total_cases'][i+6]+df2['total_cases'][i+5] + \
                       df2['total_cases'][i+4])   / (df2['total_cases'][i+3]+df2['total_cases'][i+2]+df2['total_cases'][i+1] + \
                       df2['total_cases'][i])
    else:
        for j in range(8):
            df2['RT'][i] =0
            i+=1
```

# 1.5 Кластеризация набора данных

## Будут использоваться такие модели кластеризации как Kmeans, Birch и GaussianMixture, так как позволяют работать с большими объемами данных и выбирать количество кластеров

### kmeans -

#### Он разбивает множество элементов векторного пространства на заранее известное число кластеров k. Основная идея заключается в том, что на каждой итерации перевычисляется центр масс для каждого кластера, полученного на предыдущем шаге, затем векторы разбиваются на кластеры вновь в соответствии с тем, какой из новых центров оказался ближе по выбранной метрике

### MiniBatchKMeans -

#### это вариант алгоритма KMeans, который использует мини-пакеты длясокращения времени вычислений, но при этом пытается оптимизировать ту же целевуюфункцию.

### GaussianMixture -

#### Модель смеси Гаусса (GMM) пытается найти смесь многомерных гауссовских распределений вероятностей, которые лучше всего моделируют любой входной набор данных. В простейшем случае ГММ могут использоваться для поиска кластеров таким же образом, как k-средние

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn import metrics
```

```python
# Скалирование данных
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df2)
X_scale = scaler.transform(df2)
scaled_data = pd.DataFrame(X_scale, columns=df2.columns)
```

### KMeans

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
label = kmeans.fit_predict(scaled_data)
```

```python
pd.Series(label).value_counts()
```

### MiniBatchKMeans

```python
label2 = MiniBatchKMeans(n_clusters=3).fit_predict(scaled_data)
```

```python
pd.Series(label2).value_counts()
```

### GaussianMixture

```python
gmm = GaussianMixture(3, covariance_type='full', random_state=0).fit(scaled_data)
labels3 = gmm.predict(scaled_data)
```

```python
pd.Series(labels3).value_counts()
```

### Калински-Харабаш

### Чем больше значений метрик, тем лучше расположены кластеры, следовательно далее будет использоваться kmeans

```python
print('метрики kmeans = ' + str(metrics.calinski_harabasz_score(scaled_data, label)) + '\n' +
     'метрики Minibatch = ' + str(metrics.calinski_harabasz_score(scaled_data, label2)) + '\n' +
      'метрики GaussianMixture = ' + str(metrics.calinski_harabasz_score(scaled_data, labels3)))
```

### Индекс Дэвиса-Булдина

### Этот показатель измеряет сходство ваших кластеров, что означает, что чем ниже оценка, тем лучше разделение между вашими кластерами

```python
from sklearn.metrics import davies_bouldin_score
```

```python
print('метрики kmeans = ' + str(davies_bouldin_score(scaled_data, label)) + '\n' +
     'метрики Minibatch = ' + str(davies_bouldin_score(scaled_data, label2)) + '\n' +
      'метрики GaussianMixture = ' + str(davies_bouldin_score(scaled_data, labels3)))
```

### Делая вывод по этим метрикам, можно сказать, что Minibatch и kmeans отработали +- одинаково хорошо. Поэтому для дальнейшей работы будет использоваться kmeans

### График Kmeans

```python
# График распределения кластеров в kmeans
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=label,cmap='rainbow', alpha=0.4)
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```
